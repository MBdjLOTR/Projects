```python
import cv2
import numpy as np

def calculate_negative_space_percentage(image_path):
    # Load the image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Threshold the image to binary (black and white)
    _, binary_image = cv2.threshold(image, 240, 255, cv2.THRESH_BINARY)

    # Get the total number of pixels
    total_pixels = binary_image.size

    # Count the number of white (background) pixels
    white_pixels = np.sum(binary_image == 255)

    # Calculate the percentage of negative space
    negative_space_percentage = (white_pixels / total_pixels) * 100

    return negative_space_percentage

# Example usage
image_path ='C:/Users/Medhavi/Documents/medhavi/green energy.jpg'  # Replace with your image path
percentage = calculate_negative_space_percentage(image_path)
print(f"The percentage of negative space in the image is {percentage:.2f}%")
```
